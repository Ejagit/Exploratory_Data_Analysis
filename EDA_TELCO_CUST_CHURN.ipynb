4. Berapa rata-rata TOTAL CHARGES dari customer NON-SENIOR CITIZEN (Senior Citizen = 0) yang memiliki PHONE SERVICE dan memiliki INTERNET SERVICE FIBER OPTIC?

6. Top 10 customer yang memiliki total charges paling sedikit?

```python
#Bagi customer yang TIDAK MEMILIKI PARTNER dan TIDAK MEMILIKI DEPENDENTS, berapa rata-rata dari tenure mereka?
syarat_1 = df['Partner'] == 'No'
syarat_2 = df['Dependents']== 'No'
df1 = df[syarat_1 & syarat_2]
df1
```

```python
df1['tenure'].mean()
```

```python
#Bagi customer yang MEMILIKI PARTNER dan TIDAK MEMILIKI DEPENDENTS, berapa maksimal MonthlyCharges yang tercatat?
syarat_3 = df['Partner'] == 'Yes'
syarat_4 = df['Dependents'] == 'No'
df2 = df[syarat_3 & syarat_4]
df2.shape
```

```python
df2['MonthlyCharges'].max()
```

```python
#Berapa banyak customer yang TIDAK MEMILIKI PARTNER namun MEMILIKI DEPENDENTS?
syarat_5 = df['Partner'] == 'No'
syarat_6 = df['Dependents'] == 'Yes'
df3 = df[syarat_5 & syarat_6 ]
df3.shape
```

```python
#Berapa rata-rata TOTAL CHARGES dari customer NON-SENIOR CITIZEN (Senior Citizen = 0) yang memiliki PHONE SERVICE dan memiliki INTERNET SERVICE FIBER OPTIC?
syarat_7 = df['SeniorCitizen'] == 0
syarat_8 = df['PhoneService'] == 'Yes'
syarat_9 = df['InternetService'] == 'Fiber optic'
df4 = df[syarat_7 & syarat_8 & syarat_9]
df4.shape
```

```python
df4['TotalCharges'].mean()
```

```python
#Top 5 customer yang memiliki total charges paling banyak?
top10_cust = (df
            .groupby('customerID') 
            .agg(num_trans=('TotalCharges','sum'))
            .reset_index()
            .sort_values('num_trans',ascending=False)
            .head(10)
            )
top10_cust
```

```python
plt.figure(figsize=(8,6))
top10_cust['customerID'] = top10_cust['customerID'].astype(str)
sns.barplot(data=top10_cust, x='customerID', y='num_trans')
plt.xticks(rotation=90)
plt.xlabel('Customer ID')
plt.ylabel('Total Charges')
plt.title('Top 10 Customers by Total Charges')
plt.show()
```

```python
#Top 10 customer yang memiliki total charges paling sedikit?
low10_cust = (df
            .groupby('customerID') 
            .agg(num_trans=('TotalCharges','sum'))
            .reset_index()
            .sort_values('num_trans',ascending=False)
            .tail(10)
            )
low10_cust
```

```python
plt.figure(figsize=(8,6))
top10_cust['customerID'] = low10_cust['customerID'].astype(str)
sns.barplot(data=top10_cust, x='customerID', y='num_trans')
plt.xticks(rotation=90)
plt.xlabel('Customer ID')
plt.ylabel('Total Charges')
plt.title('Low 10 Customers by Total Charges')
plt.show()
```
